```python
import pandas as pd
from uuid import uuid4
import csv
```

```python
iot_data = pd.read_csv("../iot_data/clean_iot_data.csv")
iot_data.head()
```

```python
device_1 = iot_data[iot_data['device'] == 'b8:27:eb:bf:9d:51']
device_1.head()
```

```python
device_2 = iot_data[iot_data['device'] == '00:0f:00:70:91:0a']
device_2.head()
```

```python
device_3 = iot_data[iot_data['device'] == '1c:bf:ce:15:ec:4d']
device_3.head()
```

```python
def create_sql(df: pd.DataFrame):
    temp = "INSERT INTO sensors VALUES "
    for _, row in df.iterrows():
        timestamp = row['ts'][:10]
        timestamp += " "
        timestamp += row['ts'][11:19]

        light = 0
        if row['light']:
            light = 1

        motion = 0
        if row['motion']:
            motion = 1

        val = f"('{str(uuid4())}','{timestamp}','{row['device']}',{row['co']},{row['humidity']},{row['temp']},{row['lpg']},{row['smoke']},{light},{motion}),\n"

        temp += val
    temp += ';'

    return temp
```

```python
def create_csv(df: pd.DataFrame, path: str):
    df.to_csv(path, index=False, quoting=csv.QUOTE_STRINGS)
```

```python
# export sql dump for each devices
dev1_sql = create_sql(device_1)
device1_file = open('device1.sql', 'w')
device1_file.write(dev1_sql)

dev2_sql = create_sql(device_2)
device1_file = open('device2.sql', 'w')
device1_file.write(dev2_sql)

dev3_sql = create_sql(device_3)
device1_file = open('device3.sql', 'w')
device1_file.write(dev3_sql)
```

```python
# convert clean data csv to sql format
iot_sql = create_sql(iot_data)
sql_file = open('sql/case4.sql', 'w')
sql_file.write(iot_sql)
```

```python
# separate clean iot data to each device and convert to csv
create_csv(device_1, "device_1.csv")
create_csv(device_2, "device_2.csv")
create_csv(device_3, "device_3.csv")
```
